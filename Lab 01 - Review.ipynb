### Notebook cells

Jupyter notebooks are made of cells, which can be either code or text formatted using [Markdown](https://www.markdownguide.org/basic-syntax/).  The type of the current cell is shown in the dropdown menu on the toolbar at the top of the screen, and you can change it there.

To **run** a code cell or format a Markdown cell, click the run button on the toolbar or press **shift-Enter**.  To run a cell without moving down to the next one, press **control-Enter**.

To **insert** a cell, use the Insert menu.  To **split** a cell, use Split Cell in the Edit menu.

Jupyter will automatically display the value returned by the last line of a code cell, unless you end it with a semicolon: `;`

### Kernel

Your Python code is run by an IPython **kernel**.  The kernel holds values of any variables you've defined, and references to any functions you've defined or modules you've imported.  You can't directly see these values, but you can always find out what is in a variable with a `print` statement or putting the variable name alone on the last line of a code cell and running it.

Sometimes this **hidden state** can make your code do different things from what you are thinking.  You can always clear all that state by restarting the kernel with the <i class="fa-repeat fa"></i> button.  This will clear all your variables and functions from the kernel's memory, but will not touch your code, which you can then run again without any old values getting in the way.

To make sure your code all works together, make sure it runs properly when you click **Restart & Run All** in the Kernel menu.

### Lists

Square brackets `[]` make lists in Python.  E.g. `[1, 2, 3]`, `[-1.2, 'hi', [3, 4]]`.  In this course we do a lot with NumPy arrays (see below), which are very efficient in terms of speed and memory, but can't easily be appended to and can only hold a single data type.  Many numerical methods we will see stop when they meet a termination condition after some number of iterations that is not known in advance, so appending iterations or results to a list is very helpful.  The list can easily be converted to an array if needed with `np.array(my_list_variable)`.

```python
# Examples with lists

my_list = [2]
for i in range(5):
    my_list.append(my_list[-1]**2) # Negative indexing counts back from the end, so -1 refers to the last element.

for j in my_list:
    print(j)

for i in range(len(my_list)): # A bit awkward
    print(f'Item {i} is {my_list[i]}')

for i, j in enumerate(my_list): # Much nicer
    print(f'Item {i} is {j}')

print(my_list)
print(np.array(my_list))
```

### Arrays

```python
# Examples of using numpy to create 1D arrays

import numpy as np

# Create an array by hand or using arange

a = np.array([1, 2, 3, 4, 5])
b = np.arange(1, 5)
c = np.arange(1, 5 + 1)

print(f'{a = }')
print(f'{b = }')
print(f'{c = }')
print(f'{c.dtype = }') # Returns dtype('int32'), 32-bit integers

# In this course, we almost always want floating point arrays.

# Force float data type by hand or by casting an int array:

d1 = np.array([1., 2., 3., 4., 5.]) # Just one number having a . is enough
d2 = np.array([1, 2, 3, 4, 5], dtype=float)
e = c.astype(float)

print(f'{d1 = }, {d2 = }')
print(f'{e = }')
print(f'{e.dtype = }')

# Change values using array indexing:

e[0] = 0
print(f'{e = }')

e[3:6] = 1 # Set elements 3, 4, and 5 to 1
print(f'{e = }')

e[[1, 3]] = e[[3, 1]] # Swap two elements
print(f'{e = }')
```

```python
# More examples of using numpy to create 1D arrays

# Create common arrays:

print(f'{np.zeros(5) = }')
print(f'{np.ones(5) = }')
print(f'{np.zeros_like(e) = }')
print(f'{np.ones_like(e) = }')
print(f'{np.arange(5) = }, {np.arange(5, 10) = }')
print(f'{np.linspace(0., 1., 6) = }')
print(f'{np.linspace(0., 10., 4) = }')

# Create an empty int array and fill it in by hand:

f = np.zeros(9) # By default, dtype=float
print(f'{f = }')

for i in range(1, 9):
    f[i] = f[i-1]+i

print(f'{f = }')

# Fill arrays with list comprehensions:

g = np.array([i**2 for i in range(5)])
h = np.array([i for i in range(10) if i % 2 == 1]) # i % 2 is the remainder when i is divided by 2
l = np.array([i if i % 2 == 0 else -i for i in range(10)])
print(f'{g = }')
print(f'{h = }')
print(f'{l = }')
```

Matplotlib has different ways it can work in Jupyter notebooks.  To get interactive plots rather than the default static ones, you need the line `%matplotlib widget` before you import matplotlib.pyplot. Future labs will need interactive plots.

It's easy to get mixed up which plot you're referring to when you're adding to an existing one, and if you're not careful you can end up modifying an old plot elsewhere in the notebook.  It is *highly recommended* that you:
- Keep all code for each plot in just one cell.
- Use figure and axis objects, i.e. `fig, ax = plt.subplots()` then `ax.plot`, instead of `plt.plot`.  You will often find `plt.plot` used in online examples, so be careful!

Matplotlib is very powerful, and we only use a few of its capabilities in this course.  You may find the documentation and examples useful, for example [this quick start guide](https://matplotlib.org/stable/users/explain/quick_start.html).

If you make lots of interactive plots you may start to get warnings about having too many open.  While they are interactive they use some CPU and kernel bandwidth.  You can make a plot non-interactive by clicking the blue power button in the top right.  You can have as many non-interactive plots in a notebook as you like without causing problems.

```python
# You can also nest or compose functions when you want to develop complex logic step by step

import math #for factorial

def myapproxsin(x, num_terms=5):
    '''Approximate sin(x) by hand using a truncated Taylor series polynomial.

    Args:
        x (float): x
        num_terms (int): number of terms to use from the truncated Taylor series polynomial

    Returns:
        float: approximation of sin(x)
    '''
    
    total = 0
    
    for i in range(0, num_terms):
        term = x ** (2 * i + 1) / math.factorial(2 * i + 1)
        if i % 2 == 0:
            total = total + term
        else:
            total = total - term
            
    return total

def myapproxcos(x, num_terms=5):
    '''Approximate sin(x) by hand using a truncated Taylor series polynomial.

    Args:
        x (float): x
        num_terms (int): number of terms to use from the truncated Taylor series polynomial

    Returns:
        float: approximation of sin(x)
    '''
    
    total = 0
    
    for i in range(0, num_terms):
        term = x ** (2 * i) / math.factorial(2 * i)
        if i % 2 == 0:
            total = total + term
        else:
            total = total - term
            
    return total

def myapproxtan(x, num_terms=5):
    '''Approximate sin(x) by hand using a truncated Taylor series polynomial.

    Args:
        x (float): x
        num_terms (int): number of terms to use from the truncated Taylor series polynomial

    Returns:
        float: approximation of sin(x)
    '''
    
    return myapproxsin(x, num_terms) / myapproxcos(x, num_terms)

x = np.pi

print(f'{myapproxsin(x) = }')
print(f'{myapproxsin(x, 10) = }')
print(f'{myapproxsin(x, 15) = }')
print(f'{np.sin(x) = }')
print('')
print(f'{myapproxcos(x) = }')
print(f'{myapproxcos(x, 10) = }')
print(f'{myapproxcos(x, 15) = }')
print(f'{np.cos(x) = }')
print('')
print(f'{myapproxtan(x) = }')
print(f'{myapproxtan(x, 10) = }')
print(f'{myapproxtan(x, 15) = }')
print(f'{np.tan(x) = }')
```

```python
# You can also use optional keyword arguments to change how a function works

def myfullname(first, last, style = 'default') -> str:
    '''Format first and last name into a full name, choosing from one of several styles.

    Args:
        first (str): first name
        last (str): last name
        style (str): one of 'default', 'compact', or 'formal'

    Returns:
        str: full name
    '''

    if style == 'default':
        return f'{first} {last}'

    if style == 'compact':
        return f'{first} {last[0]}'

    if style == 'formal':
        return f'{last}, {first}'

    return f'{first} {last}'

print(myfullname('James', 'Williams'))
print(myfullname('James', 'Williams', style='default'))
print(myfullname('James', 'Williams', style='compact'))
print(myfullname('James', 'Williams', style='formal'))
print(myfullname('James', 'Williams', style='invalid'))
```

```python
# Some numerical methods in this course involve a lot of iterations, and you may not want to print all of them.
# Here is a recipe for printing a counter every power of 2, which can be useful for tracking progress.

import time
from datetime import datetime

n = 100

for i in range(0, n):
    # Print iteration only if the counter is a power of 2 or the final value:
    if (i & (i - 1) == 0) or i == n - 1: # Look up binary numbers and bitwise operators to understand i & (i - 1) == 0
        print(f'{i = :>{2}d}, time = {datetime.now()}') #>{2} means left-pad with spaces to a width of 2 characters

    # Pretend to do something time-consuming:
    time.sleep(0.05)
```
